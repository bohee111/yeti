```python
'''
국민여행조사 활용도 부분을 어필할 떄 쓰면 좋을 것 같은 내용:

저희는 국민여행조사 데이터를 활용하여 고객의 '비용 민감도'를 분석했습니다.
이 데이터 분석을 통해 고객의 구매 시기, 여행지 선택 이유, 총 경비, 월 소득이 민감도에 미치는 영향을 파악하고,
이를 기반으로 '비용 고민감자'와 '저민감자'를 분류하는 규칙 기반 지표를 수립했습니다.
국민여행조사 데이터에 이 지표를 적용한 결과, 현재 고객군의 X%가 고민감 그룹으로, Y%가 저민감 그룹으로 분류되는 것을 확인했습니다.
이러한 분류는 향후 맞춤형 마케팅 및 상품 개발의 기초 자료가 될 것입니다."
'''
```

```python
#파일 불러오기

import pandas as pd
# 경로 및 정확한 파일명 설정
file_path = "/Users/ubi/Downloads/무제 폴더 2/2024년 국민여행조사_해외여행_RAWDATA.xlsx"

# 엑셀 파일 읽기
df = pd.read_excel(file_path, sheet_name=0)

# 확인
print(df.shape)
df.head()
```

```python
#컬럼명 확인
print(df.columns.tolist())
```

```python
#방문 국가-일본 필터링

japan_country_columns = [
    '해외] 방문국가1',
    '해외] 방문국가2',
    '해외] 방문국가3',
    '해외] 방문국가4',
    '해외] 방문국가5'
]

# 값이 157인 셀이 하나라도 있는 행 필터링
df_japan = df[df[japan_country_columns].isin([157]).any(axis=1)].copy()

df_japan.head()
```

```python
#구매 시기/여행지 선택 이유/에상 총 경비/개인 월 평균 소득 참고한 비용 '저비용 추구자'/'편의성 추구자' 분류
```

```python
#규칙1: 구매 시기
'''
1. 6개월 이상 전에 2. 4~6개월 전에 3. 2~3개월 전에 4. 1개월 전에
5. 2~3주 전에 6. 4~7일 전에 7. 2~3일 전에 8. 하루 전에 9. 당일

이떄, 즉, 여행 임박 구매 시점의 고객은 선택의 여지가 없고, 가격이 오를 가능성이 높으며,
충분한 비교 없이 예산을 초과할 수 있다는 불안감을 더 크게 느낄 것이고
이런 요인들이 복합적으로 작용하여,
긴급할수록 가격에 대한 민감도가 더 높아진다고 판단
''''
```

```python
#항목별 비율 확인
distribution = df_japan['AA.[해외] Q1-1. 여행사 상품 구매시기'].dropna().value_counts(normalize=True).sort_index()
print(distribution.apply(lambda x: f"{x:.1%}"))
```

```python
#=> 3,4 번을 선택한 비중이 약 75%이므로 중립(3, 4) 이전(저민감자) 이후(고민감자)로 판단
```

```python
# 분류 진행
import numpy as np
conditions_time = [
    df_japan['AA.[해외] Q1-1. 여행사 상품 구매시기'].isin([5, 6, 7, 8, 9]),
    df_japan['AA.[해외] Q1-1. 여행사 상품 구매시기'].isin([1, 2])
]
choices_time = [1, -1]
df_japan['score_time'] = np.select(conditions_time, choices_time, default=0)
```

```python
#규칙2: 여행지 선택 이유(1순위, 2순위, 3순위)
'''
1. 여행지 지명도 2. 볼거리 제공
3. 저렴한 여행경비 4. 이동 거리
5. 여행할수있는시간 6. 숙박시설
7. 쇼핑
8. 음식
9. 교통편
10. 체험 프로그램 유무 11. 경험자의 추천
12. 관광지 편의시설 13. 교육성
14. 여행 동반자 유형 15. 기타

이때, '여행지 선택 이유'라는 직접적인 정보인만큼 1/2/3순위에 따라 3/1/2점 점수 차등적으로 부여
'''
```

```python
REASON_CHEAP_TRAVEL = 3

conditions_reason = [
    # 조건 1: 1순위가 '저렴한 여행경비'인 경우 (+3점)
    df_japan['AA.[해외] Q3. 여행지 선택 이유_1순위'] == REASON_CHEAP_TRAVEL,

    # 조건 2: 2순위가 '저렴한 여행경비'인 경우 (+2점)
    # (주의: 1순위가 3이 아닌 경우에만 이 조건이 검토됨, np.select의 특성)
    df_japan['AA.[해외] Q3. 여행지 선택 이유_2순위'] == REASON_CHEAP_TRAVEL,

    # 조건 3: 3순위가 '저렴한 여행경비'인 경우 (+1점)
    # (주의: 1, 2순위가 3이 아닌 경우에만 이 조건이 검토됨)
    df_japan['AA.[해외] Q3. 여행지 선택 이유_3순위'] == REASON_CHEAP_TRAVEL
]

choices_reason = [
    3,  # 첫 번째 조건 (1순위 == 3)이 True일 때 부여할 점수
    2,  # 두 번째 조건 (2순위 == 3)이 True일 때 부여할 점수
    1   # 세 번째 조건 (3순위 == 3)이 True일 때 부여할 점수
]

df_japan['score_reason'] = np.select(conditions_reason, choices_reason, default=0)
```

```python
#규칙3: 예상하는 여행 총 경비
'''
이때, 국민 여행 조사 데이터에서 경비는 ''?????
```

```python
#항목 나누기 및 항목별 비율 확인
bins = [0, 300000, 500000, 700000, 1000000, 1500000, float('inf')]
labels = ['30만원 미만', '30-50만원', '50-70만원', '70-100만원', '100-150만원', '150만원 이상']
binned_costs = pd.cut(df_japan['해외]  B_여행1차_여행 총경비'].dropna(), bins=bins, labels=labels, right=False)
distribution = binned_costs.value_counts(normalize=True).sort_index()
print(distribution.apply(lambda x: f"{x:.1%}"))
```

```python
#=> 전체에서 54%를 차지하는 150만원 이상은 저민감, 70만원 이하는 고민감으로 판단
```

```python
#분류 진행
conditions_cost = [
    df_japan['해외]  B_여행1차_여행 총경비'] < 700000,   # 70만원 미만은 고민감 (+1)
    df_japan['해외]  B_여행1차_여행 총경비'] >= 1500000  # 150만원 이상은 저민감 (-1)
]
choices_cost = [1, -1]
df_japan['score_cost'] = np.select(conditions_cost, choices_cost, default=0) # 그 외 (70만원 이상 ~ 150만원 미만)는 중립 (0)
```

```python
#규칙4: 개인 월 평균 소득
'''
1. 소득없음
2. 월평균 100만원 미만
3. 월평균 100~200만원 미만 4. 월평균 200~300만원 미만
5. 월평균 300~400만원 미만 6. 월평균 400~500만원 미만 7. 월평균 500~600만원 미만
8. 월평균 600~700만원 미만 9. 월평균 700~800만원 미만 10. 월평균800~900만원미만
11. 월평균900~1,000만원미만 12. 월평균 1,000만원 이상
'''
```

```python
# 항목별 비중 확인
income_mapping = {
    1: '소득 없음',
    2: '100만원 미만',
    3: '100-200만원',
    4: '200-300만원',
    5: '300-400만원',
    6: '400-500만원',
    7: '500-600만원',
    8: '600-700만원',
    9: '700-800만원',
    10: '800-900만원',
    11: '900-1,000만원',
    12: '1,000만원 이상'
}
distribution = df_japan['DQ6. 월 평균 본인 소득'].dropna().value_counts(normalize=True).sort_index()
distribution.index = distribution.index.map(income_mapping)
print(distribution.apply(lambda x: str(round(x * 100, 1)) + '%'))
```

```python
#=> 4,5,6,7,8 항목 비중이 약 41%이므로 이를 기준으로 고민감(1,2,3) 저민감(9,10,11,12)로 판단
```

```python
#분류 진행
conditions_income = [
    df_japan['DQ6. 월 평균 본인 소득'].isin([1, 2, 3]),
    df_japan['DQ6. 월 평균 본인 소득'].isin([9, 10, 11, 12])
]
choices_income = [1, -1]
df_japan['score_income'] = np.select(conditions_income, choices_income, default=0)
```

```python
#개별 점수 합산하여 최종 점수 계산
score_cols = ['score_time', 'score_reason', 'score_income']
df_japan['sensitivity_score'] = df_japan[score_cols].sum(axis=1)
```
